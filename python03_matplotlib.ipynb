# Matplotlib Intro


*   Useful for making plot




```python
#importing libs...
import matplotlib.pyplot as plt
import numpy as np
```

**Plotting Data**

```python
languages = ['Bengali','French','English', 'Spanish','Dutch']
people = [100,50,150,40,70]
plt.plot(languages,people)
plt.xlabel('lang')
plt.ylabel('people')
plt.title('Lang People')
plt.show()
```

**Plotting triangular data**

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z= np.cos(x)
```

```python
print(x)
```

```python
print(y)
print(z)
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

Adding labels of axis

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

 **Parabola**

The equation of parabola is y=x^2


```python
#parabola
x=np.linspace(-10,10,20)
y = x**2

plt.plot(x,y)
plt.show()
```

Sin and Cos value plot

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x), 'g-') #x is the angle and y is the sine value
plt.plot(x,np.cos(x),'r--')
plt.show()
```

**Bar Plot**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Bengali','French','English', 'Spanish','Dutch']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('lang')
plt.ylabel('people')
plt.title('Lang People')
plt.show()
```

**Pie Chart**

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1]) #shapes a rectangular figure
languages = ['Bengali','French','English', 'Spanish','Dutch']
people = [100,50,150,40,70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

**Scatter Plot**

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax= fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

**3D Scatter Plot**

```python
fig3 = plt.figure()
plt.axes(projection = '3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
