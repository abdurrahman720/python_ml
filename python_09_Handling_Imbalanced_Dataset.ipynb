# Imbalanced Dataset

Some datasets have the target values but they are not balanced. For example we have a dataset of credit card transactions that predict if the transaction was successful or not. (label is 0 and 1)

Now , the in dataset, there's an issue. The label 0 has more than 200k+ counts where 1 has 400 counts.

This kind of dataset is not suitable for trainning our model. We should only take 400 counts of label 0 also.

````python
import pandas as pd
````

````python
card_data = pd.read_csv('/content/credit_data.csv')
card_data.head()
````

````python
#let's check the target column : 'class'
card_data['Class'].value_counts()
````

In the result above, we see the legit transaction (labeled as 0) is too much where the fraud transaction (labeled as 1) is a few.
With this dataset, our model won't able to analyze the data properly.

````python
#separating the legit and fraudulent transaction
legit = card_data[card_data.Class==0]
fraud = card_data[card_data.Class==1]

print(legit.shape)
print(fraud.shape)
````

**Under Sampling**

Build a new dataset containing similar distribution of Legit and Fraudlent Transactions

*Number of fraudlent transaction = 86*

````python
legit_sample = legit.sample(n=86)
print(legit_sample.shape)
````

Now, we need to concatate the new legit_sample and fraud in a new dataset



```
pd.concat(['the sample field to be placed first','the opposite target field'], axis= '0 if row wise')
```



````python
new_dataset = pd.concat([legit_sample, fraud], axis=0)

new_dataset.head()
````
