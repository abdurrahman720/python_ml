# Label Encoding


*  Label encoding is also a data pre-processing method.

Supposed, we have the target data labeled 'male' and 'female'. To handle the model more efficiently, we need to encode the 'male' and 'female' as numerical value.




```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

**Label encoding of breast cancer data frame**

```python
cancer_data = pd.read_csv('/content/breast_cancer_data.csv')
cancer_data.head()
```

The dataset above, shows the target is in 'diagnose' columns where the labeles are 'M' 'B'

Let's check on sns

```python
import seaborn as sns
sns.countplot(x='diagnosis', data=cancer_data)
```

```python
# decalre the instance of label encoder
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(cancer_data.diagnosis)
print(labels) #diagnosis's label will be converted into 0 and 1
```

**0-->Bengin**
**1-->Maligant**

These numerical values are encoded on alphabetical order.

Now we need to add the targets in the dataset

```python
cancer_data['target'] = labels
cancer_data.head()
```

```python
#now compare with the previous sns . plot will be same

sns.countplot(x='target', data=cancer_data)
```
