# Seaborn
*  **Data Visualization Library**


importing library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```



*   **Note: Seaborn has some built-in data sets**




```python
#total bill vs tips dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_theme()
```

**Visualize Data**


---



*   Replot

**We are going to export a figure that shows the ratio of tips and total bill for different times.It also shows whether the customer is male or female. There is size indicator also**

```python
#visualize the data
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='sex', style='sex', size='size')
```



*   **Scatter Plot**


```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

**We are gonna find out the species from a iris dataset based on sepal length and petal length**

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
#try with sepal length and petal width
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris)
```



*  **Count Plot**

*Titanic datasets*

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

**Count total people according to class**


```python
sns.countplot(x='class',data=titanic)
```

**Predict whether a person survived or not**

```python
sns.countplot(x='alive', data=titanic)
```



*   **Bar chart**

**Getting survival statistic based one sex and class**




```python
sns.barplot(x='sex', y='survived', hue='class' , data=titanic)
```

***House Price***

```python
from sklearn.datasets import fetch_california_housing
housing  = fetch_california_housing()
print(housing) #dtype is np array
```

```python
house = pd.DataFrame(housing.data, columns=housing.feature_names) #converting into pandas df
house.head()
```

```python
house['Price'] = housing.target #adding data
house.head()
```



*   Distribution Plot




```python
sns.displot(house['Price'])
```

**Correlation**


*   Positive Correlation
*   Negative Correlation



---


**Heat Map**

If crime rate as CRIM increases then price will decrease. This is Negatively Correlative.

If RM as room increases then price will increase. This is Positively Correlative.


 The first value as relative is opposite than the absolute value.



```python
correlation = house.corr() #refer from pd df.
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f',annot=True, annot_kws={'size':8}, cmap='Blues')
```
